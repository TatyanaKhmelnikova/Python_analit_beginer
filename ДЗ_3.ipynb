### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома 
### Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated'] = ((df['yr_renovated'])-(df['yr_built']))
df.loc[(df['delta_renovated'] < 0), 'delta_renovated'] = 0
df.head()
```

```python
df[df['delta_renovated'] > 0]
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['Date_year'] = pd.to_datetime(df['date'])
df['year'] = df['Date_year'].dt.year
df['month'] = df['Date_year'].dt.month
df.head()
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns= ['date', 'zipcode', 'lat', 'long'], inplace=True)
```

# 3 задача

Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния домов

```python
group_data = df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

```python
group_data
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(
    index='waterfront',
    columns='view',
    aggfunc='count',
    values='condition'
)
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=df['bedrooms'], columns=df['floors'])
```

```python
pd.crosstab(index=df['bedrooms'], columns=df['floors']).max().sort_values().to_frame()
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index=df['condition'], columns=df['grade']).median()
```
